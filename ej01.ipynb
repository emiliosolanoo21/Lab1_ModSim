```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def campo_de_direcciones(f, xmin, xmax, ymin, ymax, xstep, ystep, field_type='F'):
    x = np.linspace(xmin, xmax, int((xmax - xmin) / xstep) + 1)
    y = np.linspace(ymin, ymax, int((ymax - ymin) / ystep) + 1)
    
    X, Y = np.meshgrid(x, y)
    
    U, V = np.zeros(X.shape), np.zeros(Y.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            U[i, j], V[i, j] = F(X[i, j], Y[i, j], f, field_type)
    
    plt.figure(figsize=(8, 6))
    
    plt.quiver(X, Y, U, V, color='#5a189a', angles='xy')
    
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.title('Campo de Direcciones')
    
    plt.show()
    
def F(x, y, f, field_type='F'):
    dy_dx = f(x, y)
    if field_type == 'F':
        return 1, dy_dx
    elif field_type == 'N':
        norm = np.sqrt(1 + dy_dx**2)
        return 1 / norm, dy_dx / norm
    else:
        raise ValueError("field_type must be 'F' or 'N'")
```

```python
def edo(x, y):
    return y - x

xmin, xmax = -10, 10
ymin, ymax = -10, 10
xstep, ystep = 1, 1

campo_de_direcciones(edo, xmin, xmax, ymin, ymax, xstep, ystep, field_type='F')
```

```python
def lineas_de_flujo(f, xmin, xmax, ymin, ymax, xstep, ystep, field_type='F'):
    x = np.linspace(xmin, xmax, int((xmax - xmin) / xstep) + 1)
    y = np.linspace(ymin, ymax, int((ymax - ymin) / ystep) + 1)
    
    X, Y = np.meshgrid(x, y)
    
    U, V = np.zeros(X.shape), np.zeros(Y.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            U[i, j], V[i, j] = F(X[i, j], Y[i, j], f, field_type)
    
    plt.figure(figsize=(8, 6))
    
    plt.quiver(X, Y, U, V, color='#00b4d8', angles='xy')
    
    plt.streamplot(X, Y, U, V, color='#5a189a')
    
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.title('Campo de Direcciones con Lneas de Flujo')
    
    plt.show()
```

```python
lineas_de_flujo(edo, xmin, xmax, ymin, ymax, xstep, ystep, field_type='N')
```
